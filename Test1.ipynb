```python
df.drop(["date","lights"],axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
print(scaled)
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(["Appliances"], axis = 1).values
y = df["Appliances"].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
model = RandomForestRegressor(random_state = 1)
model.fit(x_train, y_train)
y_try = model.predict(x_test)
mae = mean_absolute_error(y_test,y_try)
print('MAE: ',mae)
round(mae,2)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_try))
round(rmse,2)
```

```python
from sklearn.metrics import mean_squared_log_error
rmsle = np.sqrt(mean_squared_log_error(y_test, y_try))
round(rmsle,3)
```

```python
from sklearn.metrics import r2_score
COD = np.sqrt(r2_score(y_test, y_try))
round(COD,2)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha= 0.4 )
ridge.fit(x_train, y_train)
from sklearn.metrics import mean_squared_log_error
rmsle = np.sqrt(mean_squared_log_error(y_test, y_try))
round(rmsle,2)
```
